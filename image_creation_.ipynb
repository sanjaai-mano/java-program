```python
from os import pipe
from diffusers import StableDiffusionPipeline
import torch
import matplotlib.pyplot as plt

pipeline = StableDiffusionPipeline.from_pretrained("compVis/stable-diffusion-v1-4")
pipeline.to("cuda")

prompt = "polo car"
image = pipeline(prompt, num_inference_step=50).images[0]

plt.imshow(image)
plt.axis("off")
plt.show()
```
